```python
import pandas as pd
```

```python
df = pd.read_csv('/content/imigrantes_canada.csv')
```

```python
df
```

```python
df.info()
```

```python
## analisar tendencias de imigração do brasil ao longo dos anos
```

```python
## mudando o index. Inplace para que seja executado neste data frame, sem criar outro
df.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
brasil
```

```python
argentina = df.loc['Argentina', anos]
argentina
```

```python
brasil_dict = {'ano':  brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
argentina_dict = {'ano': argentina.index.tolist(), 'imigrantes': argentina.values.tolist()}
data_argentina = pd.DataFrame(argentina_dict)
data_argentina
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.plot(data_argentina['ano'], data_argentina['imigrantes'])
plt.legend(['Brasil', 'Argentina'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
## melhorando a visualização:definindo os pontos entre cada eixo
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
## para deixar aparecendo somente o grafico, sem o codigo automatico que aparece. Precisa estar na ultima linha
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
##definindo informações dos eixos, grossura da linha do gráfico, colocando marcador(marker='o') e definindo cor
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3, color='purple')
##nomeando a figura e dando as configurações da escrita
ax.set_title('Imigração do Brasil para o Canadá\n1980-2013', fontsize=18, loc='left')
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Número de imigrantes', fontsize=14)
##ajustando tamanho das letras do parametro do eixo
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
## para deixar espaçado por multiplos de 5 no eixo x
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
##adicionando grades com estilo específico
##plt.grid(linestyle='--')
##para que a linha de cima da grade suma
ax.spines['top'].set_visible(False)
##para que a linha da lateral direita suma
ax.spines['right'].set_visible(False)
##para salvar, transparent referente ao fundo, dpi resolução, bbox limites
fig.savefig('imigração_brasil_canada.png', transparent=False, dpi=300, bbox_inches='tight')
plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
#para definir espaçamento hspace = horizontal wspace = vertical
fig.subplots_adjust(hspace=0.5, wspace=0.3)
#dando titulo geral
fig.suptitle('Imigração dos 4 maiores países da América do Sul para Canadá\n1980-2013')

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colombia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

#para executar em todos os axs
for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
  ax.set_xlabel('Ano')
  ax.set_ylabel('Número de imigrantes')

##deixando todos na mesma escala de eixo y
ymin = 0
ymax = 7000
for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

```python
##aplicando filtro conforme a região desejada
america_sul = df.query('Região == "América do Sul"')
```

```python
america_sul
```

```python
##criando um novo data frame ordenando os dados do maior para o maior
##dentro dos parenteses colocar a coluna pelo qual vc quer ordenar
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
##destacando uma informação
cores = []
for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))
##grafico em barra=ax.bar // grafico em barra invertida ax.barh
##coloquei o data frame que criei depois com os valores ordenados
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do sul: Brasil foi o 4 país com mais imigrantes\npara o Canadá no período 1980-2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

##adicionando informaçõs em cada barra
for i, v in enumerate(america_sul_sorted['Total']):
##v + 20 indica a posição do texto, v = barra, i = y, str (v)=valor barra em texto
  ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')
## retirando a caixa do grafico para ficar mais limpo
ax.set_frame_on(False)
##retirando eixo x
ax.get_xaxis().set_visible(False)
##retirando marcação ao lado dos países, axis indicando eixo, which indicando quais eixos, length indicando tamanho
ax.tick_params(axis='both',which='both', length=0)
##salvando
fig.savefig('imigracao_america_sul.png', transparent=False, dpi=300, bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

ax.set_title('Países com maior imigração para Canadá\n1980-2013', loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
plt.show()
```

```python
def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=palette)
  ax.set_title('Países com maior imigração para o Canadá\n1980-2013', loc='left', fontsize=18)
  ax.set_xlabel('Número de imigrantes')
  ax.set_ylabel('')
  plt.show()
```
